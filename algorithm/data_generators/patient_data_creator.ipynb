```python
import pandas as pd
import numpy as np
import time
import random
import datetime
```

```python
id = [i for i in range(1, 101)]
central_temperature = [random.randrange(35, 41) for i in range(100)]
cardio_frequency = [random.randrange(60, 180) for i in range(100)]
resp_frequency = [random.randrange(15, 30) for i in range(100)]
total_leucocyto = [random.randrange(2000, 20000) for i in range(100)]
left_deviation = [random.choice(['0', '1']) for i in range(100)]
sistolic_arterial_pressure = [random.randrange(60, 130) for i in range(100)]
average_arterial_pressure = [random.randrange(40, 80) for i in range(100)]
oliguria = [random.uniform(0.25, 0.75) for i in range(100)]
creatinine = [random.uniform(0.5, 3.0) for i in range(100)]
pao2_fio2_relation = [random.randrange(150, 400) for i in range(100)]
spo2 = [random.choice(['0', '1']) for i in range(100)]
platelet = [random.randrange(50000, 200000) for i in range(100)]
lactate = [random.choice(['0', '1']) for i in range(100)]
consciousness_status = [random.randrange(0, 1) for i in range(100)]
bilirubin = [random.choice(['0', '1']) for i in range(100)]
user_id = [random.randrange(1, 5) for i in range(100)]
pacient_id = [i for i in range(1, 101)]
created_at = [datetime.datetime.now() for i in range(100)]
updated_at = [datetime.datetime.now() for i in range(100)]
```

```python
df = pd.DataFrame({'id' : id,
                   'central_temperature' : central_temperature,
                   'cardio_frequency' : cardio_frequency,
                   'resp_frequency' : resp_frequency,
                   'total_leucocyto' : total_leucocyto,
                   'left_deviation' : left_deviation,
                   'sistolic_arterial_pressure' : sistolic_arterial_pressure,
                   'average_arterial_pressure' : average_arterial_pressure,
                   'oliguria' : oliguria,
                   'creatinine' : creatinine,
                   'pao2_fio2_relation' : pao2_fio2_relation,
                   'spo2' : spo2,
                   'platelet' : platelet,
                   'lactate' : lactate,
                   'consciousness_status' : consciousness_status,
                   'bilirubin' : bilirubin,
                   'user_id' : user_id,
                   'pacient_id' : pacient_id,
                   'created_at' : created_at,
                   'updated_at' : updated_at})
```
